```python
import pyaudio
import numpy as np
import time
import matplotlib.pyplot as plt

from tkinter import TclError
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import keyboard

CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100
```

```python
#plt show
#basic fft

#tkinter graph
fs = 100
t = np.arange(0, 5, 1/ fs)
 
f1, f2, f3, f4 = 10, 35, 70, 3

signal = 2 * np.sin(f1 * 2 * np.pi * t) + 0.6 * np.sin(f2 * 2 * np.pi * t)\
    + 2 * np.sin(f3 * 2 * np.pi * t) + 0.6 * np.sin(f4 * 2 * np.pi * t)
    
fft = np.fft.fft(signal) / len(signal)
fft_mag = abs(fft) #y값 설정

nt = np.linspace(-(fs/2), fs/2, len(signal)) #x값 설정

# f, a = plt.subplots(2, figsize = (15,7)) # plt graph

#Showing by tk window
root = tk.Tk()
fig = plt.Figure(figsize=(5,4), dpi = 100)

#Put canvas that shows graph
ax = fig.add_subplot(1,1,1)
ax.plot(nt, 2 * np.fft.fftshift(fft_mag))
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.draw()
canvas.get_tk_widget().pack()

root.mainloop()
```

```python
#finite time fft

p = pyaudio.PyAudio()

stream = p.open(
    format= FORMAT,
    channels=CHANNELS,
    rate = RATE,
    input = True,
    frames_per_buffer=CHUNK
)

fs = 2**10 # 청크 사이즈와 동일
nt = np.linspace(-(fs/2), fs/2, CHUNK) #x값 설정

#Showing by tk window
root = tk.Tk()
fig = plt.Figure(figsize=(5,4), dpi = 100)

ax = fig.add_subplot(111)
ax.set_ylim(-1, 1)
# ax.plot(nt, nt)

canvas = FigureCanvasTkAgg(fig, master=root)
canvas.draw()
canvas.get_tk_widget().pack()


print("start")

while(True):
    data = stream.read(CHUNK)
    data_np = np.frombuffer(data, dtype=np.int16)
    
    signal = data_np/len(data_np)
    fft = np.fft.fft(signal) / len(signal)
    fft_mag = abs(np.fft.fftshift(fft))
    
    ax.cla()
    ax.plot(nt, fft_mag)
    ax.set_ylim(-0.1,0.1)
    
    if keyboard.is_pressed('q'):
        root.destroy()
        break
    
    try:
        canvas.flush_events()
        canvas.draw()
        
    except TclError:
        print("TclError")
        break
    
print("finish")

stream.stop_stream()
p.terminate()
```

```python
p = pyaudio.PyAudio()
frames = []

def getData(in_data, frame_count, time_info, status_flags):
    frames.append(in_data)
    return None, pyaudio.paContinue

stream = p.open(
    format= FORMAT,
    channels=CHANNELS,
    rate = RATE,
    input = True,
    frames_per_buffer=CHUNK,
    stream_callback= getData
)

def gen():
    while True:
        if len(frames) == 0:
            continue
        data = []
        while True:
            if len(frames) == 0:
                break
            data.append(frames.pop(0))
            
        
        yield b''.join(data)
      
fs = 2**10 # 청크 사이즈와 동일
nt = np.linspace(-(fs/2), fs/2, CHUNK) #x값 설정


isStop = False
root = tk.Tk()
fig = plt.Figure(figsize=(10,8), dpi = 100)

ax = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
# ax.plot(nt, nt)

canvas = FigureCanvasTkAgg(fig, master=root)
canvas.draw()
canvas.get_tk_widget().pack()

fr = []
for i, x in enumerate(gen()):
    fr.append(x)
    wav = np.frombuffer(b''.join(fr), np.int16)
    
    signal = wav[-RATE::]
    fft = np.fft.fft(signal) / len(signal)
    fft_mag = abs(np.fft.fftshift(fft))

    nt = np.linspace(-(fs/2), fs/2, len(signal))
    
    ax.cla()
    ax.plot(nt, fft_mag)
    ax.axis([-(fs/2), fs/2, 0, 50])
    
    ax2.cla()
    ax2.plot(wav[-CHUNK * 100:] * 0.1)
    ax2.axis([0, CHUNK * 100, -5000, 5000])
    
    if keyboard.is_pressed('space'):
        if isStop:
            isStop = False
        else:
            isStop = True
    
    if keyboard.is_pressed('q'):
        root.destroy()
        p.terminate()
        stream.stop_stream()    
        break
    
    try:
        if isStop:
            continue
        canvas.flush_events()
        canvas.draw()
        
    except TclError:
        print("TclError")
        break
    

```
